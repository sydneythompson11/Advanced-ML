```python
# importing libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from pycaret.classification import *
```

```python
# importing the data
adult = pd.read_csv("adult.csv")
adult.head()
```

# PyCaret

```python
# PyCaret full workflow on original adult data
# NOTE: PyCaret expects raw-ish data and handles preprocessing internally

from pycaret.classification import *
from sklearn.model_selection import train_test_split

# train/test split from original dataset
train_pc, test_pc = train_test_split(
    adult,
    test_size=0.2,
    random_state=42,
    stratify=adult["income"]
)

# setup experiment
clf_setup = setup(
    data=train_pc,
    target="income",
    session_id=42,
    fold_strategy="stratifiedkfold",
    fold=5,
    numeric_imputation="median",
    categorical_imputation="mode",
    normalize=False,
    verbose=False
)

# compare top models
top_models = compare_models(n_select=3)

# tune the best one
tuned_model = tune_model(top_models[0])

# blend top models
blended_model = blend_models(estimator_list=top_models)

# stack top models
stacked_model = stack_models(estimator_list=top_models)

# finalize chosen model
final_model = finalize_model(stacked_model)

# predict on held-out test set
preds = predict_model(final_model, data=test_pc)

# show results
preds.head()
```

```python
# PyCaret full workflow on original adult data
# Excluding CatBoost so tune_model() runs cleanly

from pycaret.classification import *
from sklearn.model_selection import train_test_split

# train/test split from original dataset
train_pc, test_pc = train_test_split(
    adult,
    test_size=0.2,
    random_state=42,
    stratify=adult["income"]
)

# setup experiment
clf_setup = setup(
    data=train_pc,
    target="income",
    session_id=42,
    fold_strategy="stratifiedkfold",
    fold=5,
    numeric_imputation="median",
    categorical_imputation="mode",
    normalize=False,
    verbose=False
)

# compare top models, excluding CatBoost
top_models = compare_models(
    n_select=3,
    exclude=["catboost"]
)

# tune the best model
tuned_model = tune_model(top_models[0])

# blend top models
blended_model = blend_models(estimator_list=top_models)

# stack top models
stacked_model = stack_models(estimator_list=top_models)

# finalize chosen model
final_model = finalize_model(stacked_model)

# predict on held-out test set
preds = predict_model(final_model, data=test_pc)

# show results
preds.head()
```

```python
# Show models and parameters clearly

from pprint import pprint

print("Top 3 Models Selected:")
for i, model in enumerate(top_models):
    print(f"\nModel {i+1}: {type(model).__name__}")
    pprint(model.get_params())

print("\nTuned Best Model:")
print(type(tuned_model).__name__)
pprint(tuned_model.get_params())

print("\nStacked Model Structure:")
print(stacked_model)
```

```python
# print full details of base and meta models in the stack
print("Base models in stack:\n")
for name, model in stacked_model.estimators:
    print(f"{name}:")
    print(model)
    print()
    
print("Meta model:\n")
print(stacked_model.final_estimator)
```
